```python
import pandas as pd
```

```python
# load the iris dataset
imported = load_iris()
```

```python
iris_df = pd.DataFrame(data=imported.data, columns=imported.feature_names)
iris_df['target'] = imported.target
```

```python
# define features (X) and (Y)
X = iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]
Y = iris_df['target']
```

```python
# split the data into training  and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# create a KNN classifier with k=3
knn_model = KNeighborsClassifier(n_neighbors=3)
```

```python
# train the model
knn_model.fit(X_train, Y_train)
```

```python
# make predictions on the test set
y_pred = knn_model.predict(X_test)
```

```python
# evaluate the model
accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy of KNN model: {accuracy}")
```

```python
from sklearn.metrics import classification_report
#print classification report for detailed evaluation
print(classification_report(Y_test, y_pred))
```

```python
def core(dataset, threshold):
  col_corr = set() # using set to store all the names of correlated columns for
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
        colname = corr_matrix.columns[i] # getting the name of column
        col_corr.add(colname)
  return col_corr
```

```python
corr_features = core(X_train, 0.7) # Change 'Core' to 'core'
len(set(corr_features))
```
